```python
#定义一个删除非整数次训练的权重
import re

def check_file(filePath):
	# 统计删除总数
    count = 0
    # 删除已经生成的data_*.js 答卷js文件
    for file_path, empty_list, file_name_list in os.walk(filePath):
        # file_name_list该列表是存放目标目录中所有文件名
        for file_name in file_name_list:
            # 正则匹配需要删除的文件--根据需求修改正则表达式
            if re.match(r'mask_rcnn_coco_0[0-9][0-9][0-9]', file_name):
                # 删除匹配到的文件
                if not re.match(r'mask_rcnn_coco_0[0-9][0-9]0', file_name):
                  os.remove(file_path + file_name)
                  # 每删除一个文件＋1
                  count += 1
```
